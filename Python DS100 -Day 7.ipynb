```python
# From list of list
companies=[["Apple",111],["Facebook",222],["Google",333],["Amazon",444]]
df2_companies=pd.DataFrame(companies)
print(df2_companies)
```

```python
companies2={"Apple":["iPad","Phone","a"],"Amazon":["Aws","Prime","b"]}
df4_companies=pd.DataFrame(companies2,index=["Details 1","Details 2","Details 3"])
print(df4_companies)
```
